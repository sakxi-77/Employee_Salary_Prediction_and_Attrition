```python
import numpy as np
import pandas as pd  
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score, accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_csv(r"C:\Users\Dell\Desktop\Uv Techno\Machine Learning\Project\IBM_Employee_Data.csv")
```

```python
data.head()
```

```python
data.columns
```

```python

data = data.drop(['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours'], axis=1)
```

```python
data.columns
```

```python
data.shape
```

#encoding categorical feature


```python
data.dtypes
```

```python
data['Attrition'] = data['Attrition'].map({'Yes':1, 'No':0})
data['OverTime'] = data['OverTime'].map({'Yes':1, 'No':0})
```

```python
# Label Encoding

le = LabelEncoder()

data['Gender'] = le.fit_transform(data['Gender'])
data['BusinessTravel'] = le.fit_transform(data['BusinessTravel'])
data['MaritalStatus'] = le.fit_transform(data['MaritalStatus'])
```

```python
data.head(3)
```

```python
# One-Hot Encoding : It prevents dummy variable trap and reduces extra columns with many categories like: Department, JobRole, EducationField

data = pd.get_dummies(data, columns=['Department', 'JobRole','EducationField'], drop_first=True)
```

```python
data.head(3)
```

```python
data = data.astype(int)
```

```python
data.head(3)
```

```python
data.shape
```

Defining X and Y


```python
X = data.drop(['MonthlyIncome','Attrition'],axis = 1)
Y = data['MonthlyIncome']
```

Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42
)
```

Model Building


```python
rf_model = RandomForestRegressor(random_state=42)
```

```python
rf_model.fit(X_train, y_train)
```

```python
y_pred_rf = rf_model.predict(X_test)
```

```python
r2_score(y_test, y_pred_rf)
```

```python
mean_absolute_error(y_test, y_pred_rf)
```

```python
print(X_train.columns)
```

Attrition Prediction

```python
X_attrition = data.drop(['Attrition'], axis=1)
Y_attrition = data['Attrition']
```

```python
X_train_a, X_test_a, y_train_a, y_test_a = train_test_split(
    X_attrition, Y_attrition, test_size =0.2, random_state =42
)
```

```python
classification = RandomForestClassifier(random_state=42)
```

```python
classification.fit(X_train_a, y_train_a)
```

```python
y_pred_attrition = classification.predict(X_test_a)
```

```python
accuracy_score(y_test_a, y_pred_attrition)
```

Deployement

```python
import joblib

joblib.dump(rf_model, "salary_model.pkl")
joblib.dump(classification, "attrition_model.pkl")
```
